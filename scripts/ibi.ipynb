```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from scipy.interpolate import make_interp_spline
from scipy.signal import find_peaks
```

```python
bvp = pd.read_csv("./BVP.csv").to_numpy().ravel()
```

```python
bvp
```

```python
x = np.array(range(1700))
y = bvp[300:2000]
```

```python
plt.figure(figsize=(20, 20))
plt.plot(x, y)
plt.show()
```

```python
# 47
find_peaks(y, prominence=2)
```

```python
X_Y_Spline = make_interp_spline(x, y)
```

```python
X_ = np.linspace(x.min(), x.max(), 10000)
Y_ = X_Y_Spline(X_)
```

```python
plt.figure(figsize=(20, 20))
plt.plot(X_, Y_)
plt.title("smooth curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 43
find_peaks(Y_, prominence=2)
```
