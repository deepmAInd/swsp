```python
import datetime

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from scipy.interpolate import make_interp_spline
from scipy.signal import find_peaks
```

# Load Data

```python
df_bvp = pd.read_csv("../data/BVP.csv", header=None, names=["value"])
df_eda = pd.read_csv("../data/EDA.csv", header=None, names=["value"])
df_hr = pd.read_csv("../data/HR.csv", header=None, names=["value"])
df_temp = pd.read_csv("../data/TEMP.csv", header=None, names=["value"])
```

```python
df_bvp["timestamp"] = [datetime.datetime.now() + datetime.timedelta(minutes=x) for x in df_bvp.index]
df_eda["timestamp"] = [datetime.datetime.now() + datetime.timedelta(minutes=x) for x in df_eda.index]
df_hr["timestamp"] = [datetime.datetime.now() + datetime.timedelta(minutes=x) for x in df_hr.index]
df_temp["timestamp"] = [datetime.datetime.now() + datetime.timedelta(minutes=x) for x in df_temp.index]
```

```python
df_bvp.set_index("timestamp", inplace=True)
df_eda.set_index("timestamp", inplace=True)
df_hr.set_index("timestamp", inplace=True)
df_temp.set_index("timestamp", inplace=True)
```

```python
df_bvp.to_csv("./bvp.csv")
df_eda.to_csv("./eda.csv")
df_hr.to_csv("./hr.csv")
df_temp.to_csv("./temp.csv")
```

```python
df_bvp.to_json("./bvp.json")
df_eda.to_json("./eda.json")
df_hr.to_json("./hr.json")
df_temp.to_json("./temp.json")
```

```python
FIG_SIZE = (16, 9)
RANGE = 2000
```

The `.plot` function takes two crusial parameters `scalex` and `scaley`. In the case of the raw data you can plot the measurment of the feature on `scaley` and the point of time in `scalex`. In this case the `scalex` is the timestamp.

```python
plt.figure(figsize=FIG_SIZE)
plt.plot(df_bvp.index[1:RANGE], df_bvp.value[1:RANGE])
plt.show()
```
